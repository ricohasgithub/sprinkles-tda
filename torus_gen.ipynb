```python
import gudhi as gd
import numpy as np

# Generating a point cloud for a torus
def sample_torus(n_samples=1000, c=2, a=1):
    """
    Sample n points uniformly from a torus embedded in 3d.
    The torus is parametrized by (c, a) where c is the distance from the center of the tube to the center of the torus,
    and a is the radius of the tube.
    """
    theta = 2 * np.pi * np.random.rand(n_samples)
    phi = 2 * np.pi * np.random.rand(n_samples)
    x = (c + a * np.cos(theta)) * np.cos(phi)
    y = (c + a * np.cos(theta)) * np.sin(phi)
    z = a * np.sin(theta)

    return np.column_stack([x, y, z])

# Create a point cloud from the torus
point_cloud = sample_torus()

# Create a Rips complex from the point cloud
rips_complex = gd.RipsComplex(points=point_cloud, max_edge_length=0.5)

# Create a simplex tree from the Rips complex
simplex_tree = rips_complex.create_simplex_tree(max_dimension=2)

# Compute persistence
diagram = simplex_tree.persistence()

# Plot the persistence diagram
gd.plot_persistence_diagram(diagram)
```
